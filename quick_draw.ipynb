```python
import csv
import json
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Pre-processing and Evaluation for Quick, Draw! dataset

```python
ndjson_file = "raw_data\\full_simplified_castle.ndjson"
csv_file = "data\castle.csv"

data = []
with open(ndjson_file, "r") as f:
    for line in f:
        data.append(json.loads(line.strip()))
headers = data[0].keys()

with open(csv_file, "w", newline="") as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=headers)
    writer.writeheader()
    writer.writerows(data)

print(f"Converted NDJSON to CSV and saved as '{csv_file}'.")
```

```python
castle = pd.read_csv("data\castle.csv")
castle
```

```python
null_counts = castle.isnull().sum()
duplicate_count = castle.duplicated().sum()
duplicates = castle[castle.duplicated()]

print("Null Values in Each Column:", null_counts)
print("\nTotal Number of Duplicate Rows:", duplicate_count)
if not duplicates.empty:
    print("Duplicate Rows:")
    print(duplicates)
else:
    print("No duplicate rows found.")

castle = castle.drop_duplicates()
castle = castle.dropna()
```

### Making full img file

```python
input_files = ["data\castle.csv", "data\dragon.csv", "data\hourglass.csv", "data\knife.csv", "data\skull.csv", "data\sword.csv"]
output_csv = "img_data.csv"

header_written = False
for input_csv in input_files:
    with open(input_csv, "r") as csvfile:
        reader = list(csv.DictReader(csvfile))

    selected_rows = random.sample(reader, min(200, len(reader)))
    fieldnames = reader[0].keys()

    with open(output_csv, "a", newline="") as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        if not header_written:
            writer.writeheader()
            header_written = True
        writer.writerows(selected_rows)

print(f"Randomly selected rows from multiple files saved to '{output_csv}'.")
```

```python
img = pd.read_csv("img_data.csv")
null_counts = img.isnull().sum()
duplicate_count = img.duplicated().sum()
duplicates = img[img.duplicated()]

print("Null Values in Each Column:", null_counts)
print("\nTotal Number of Duplicate Rows:", duplicate_count)
if not duplicates.empty:
    print("Duplicate Rows:")
    print(duplicates)
else:
    print("No duplicate rows found.")

img = img.drop_duplicates()
img = img.dropna()
```
